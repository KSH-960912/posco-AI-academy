```python
from copy import deepcopy

class Set:
    def __init__(self, lst):
        self.lst = []
        for elem in lst:
            if elem not in self.lst:
                self.lst.append(elem)
        
    def add(self, elem):
        self.lst.append(elem)

    def discard(self, elem):
        self.lst.remove(elem)

    def clear(self):
        self.lst.clear()

    def __len__(self):
        return len(self.lst)
    
    def __str__(self):
        result = '{' + ', '.join(str(elem) for elem in self.lst) + '}'
        return result
    
    def __contains__(self, elem):
        if elem in self.lst:
            return True
        else:
            return False
    
    def __le__(self, other):
        for elem in self.lst:
            if elem not in other.lst:
                return False
        return True
    
    def __ge__(self, other):
        for elem in other.lst:
            if elem not in self.lst:
                return False
        return True
    
    def __or__(self, other):
        new_set = Set([])
        new_set.lst = deepcopy(self.lst)
        for elem in other.lst:
            if elem not in new_set.lst:
                new_set.lst.append(elem)
        return new_set
    
    def __and__(self, other):
        new_set = Set([])
        for elem in self.lst:
            if elem in other.lst:
                new_set.lst.append(elem)
        return new_set

    def __sub__(self, other):
        new_set = Set([])
        new_set.lst = deepcopy(self.lst)
        for elem in other.lst:
            if elem in self.lst:
                new_set.lst.remove(elem)
        return new_set
    
    def __ior__(self, other):
        self.lst = (self | other).lst
        return self

    def __iand__(self, other):
        self.lst = (self & other).lst
        return self
    
    def __isub__(self, other):
        self.lst = (self - other).lst
        return self
```

```python
a = Set([1, 2, 3, 4])
b = Set([1, 2, 3, 4])
print(a)
print(b)
print()

a.discard(4)
b.discard(1)
print(a)
print(b)
print()

print(len(a))
print(1 in a)
print(1 in b)
print()

print(a | b)
print(a & b)
print(a - b)
print()

print(a <= b)
print(a <= a | b)
print(a >= b)
print(a >= a & b)
print()

b.clear()
print(b)
```

```python
a = Set([1, 2, 3])
b = Set([3, 4])
adress_a = id(a)
a |= b
print(a)
print(adress_a == id(a))
print()

a = Set([1, 2, 3])
b = Set([3, 4])
adress_a = id(a)
a &= b
print(a)
print(adress_a == id(a))
print()

a = Set([1, 2, 3])
b = Set([3, 4])
adress_a = id(a)
a -= b
print(a)
print(adress_a == id(a))
print()
```
