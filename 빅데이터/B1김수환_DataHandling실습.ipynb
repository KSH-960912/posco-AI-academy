## 실습 4

```python
import numpy as np
from numpy.random import randn

np.random.seed(1234)
df = pd.DataFrame(randn(10, 4), columns = ['A','B','C','D'])
df['B'] # B열 선택 
df[['B','D']] # B, D열 선택
df[4:9] # 4~8행 선택
df.loc[2:8] # 2~8행
df.loc[:,['A','C']] # A,C열 선택
df.loc[3:8] #3~8행 선택
df.loc[3:3] # 3행 선택
df.iloc[0:7,[0]] # 0~6행, A열 선택
df.iloc[:,[0]] # A열선택
df.iloc[2:7] # 2~6행 선택
df.iloc[5:6] # 5행 선택(단일행)
df.query('D > 0.5') # D > 0.5 선택
df[df.B > df.D] # slice 표현으로 B > D 데이터 선택
df.query('A > 0.5 & D < 0.0') # A > 0.5 그리고 D < 0.0 데이터 
df.query('A > 0.5 | D < 0.0') # A > 0.5 혹은 D < 0.0 데이터 
```
